```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# equally divide [-1,1] to 100 points
x = np.linspace(-1, 1, 100)

f = lambda t: t**2
g = lambda t: 2**t

plt.plot(x, f(x))
plt.plot(x, g(x))

plt.grid()
```

```python
import math
x = np.linspace(0, math.pi/2)

sincube = lambda t: np.sin(t)**3
coscube = lambda t: np.cos(t)**3

plt.plot(coscube(x), sincube(x))
```
